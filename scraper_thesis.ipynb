```python
import snscrape.modules.twitter as sntwitter
import pandas as pd
from tqdm.notebook import tqdm
```

```python
tweets = []

# Define the keyword and language to be used in the query
keyword = "start-ups "
language = 'en'

#Create days that you want to scrape(manipulate date as string)
#Create a list of all dates
ds_30 = ["01", "02", "03", "04", "05", "06", "07", "08", "09"] + [str(i) for i in range(10, 31)]
ds_31 = ds_30 + ["31"]
#
days = ["2022-01-" + ds_30[i] for i in range(len(ds_30))] #September
days = days + ["2022-02-" + ds_31[i] for i in range(len(ds_31))]  #October
days = days + ["2022-03-" + ds_30[i] for i in range(len(ds_30))]  #November
days = days + ["2022-04-" + ds_31[i] for i in range(len(ds_31))] + ["2023-01-01"]  #December , including the last day of December
                                                                                   # by adding the first day of January.

# Initialize a counter to keep track of the total number of tweets scraped
for i, day in enumerate(days[:-1]):
    print("Now at day", day)
    tweet_per_day_count = 0
    query = keyword + " since:" + day + " until:" + days[i+1] + " lang:" + language
    for tweet in tqdm(sntwitter.TwitterSearchScraper(query).get_items()):
        if tweet_per_day_count > 333:
            break
        tweets.append([tweet.date, tweet.content, tweet.user.username, tweet.likeCount, tweet.user.followersCount, tweet.id, tweet.quoteCount, tweet.replyCount, tweet.retweetCount])
        tweet_per_day_count += 1
```

```python
df = pd.DataFrame(tweets, columns=['Date', 'Tweet', "Username", "Likes","Users Followers","Users id","Tweet quote", "Tweet replies", "Tweet retweets"])
print(df)
```
